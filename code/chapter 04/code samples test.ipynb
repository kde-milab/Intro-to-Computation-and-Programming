```python
def apply_to_each(L, f):
   """L жагсаалт, f функцаар
      L жагсаалтын элемент e бүрийг f(e) гэж хувьгана"""
   for i in range(len(L)):
      L[i] = f(L[i])
      
L = [1, -2, 3.33]
print('L =', L)
print('Apply abs to each element of L.')
apply_to_each(L, abs)
print('L =', L)
print('Apply int to each element of', L)
apply_to_each(L, int)
print('L =', L)
print('Apply squaring to each element of', L)
apply_to_each(L, lambda x: x**2)
print('L =', L)
```

```python
def custom_lambda(x, y):
    print(f"Raising {x} to the power of {y}")
    return x**y
def f(L1, L2):
    """L1, L2 lists of same length of numbers
    returns the sum of raising each element in L1
    to the power of the element at the same index in L2
    For example, f([1,2], [2,3]) returns 9"""
    if len(L1) != len(L2):
        raise ValueError("L1 and L2 must have the same length")
    # return sum(map(lambda a, b: a ** b, L1, L2))
    return sum(map(lambda a, b: custom_lambda(a, b), L1, L2))

# quick checks (uses existing L1, L2 if present in the notebook)
print(f([1, 2], [2, 3]))  # 9
# print(f(L1, L2))
```

```python
def get_min(d: dict) -> int:
    """Return the value in d whose key occurs first in the alphabet.
    Raises ValueError if d is empty.
    Example: {'x': 11, 'b': 12} -> 12
    """
    if not d:
        raise ValueError("d must be non-empty")
    return d[min(d)]

d = {'a': 11, 'b': 12}
print(get_min(d))
print(min(d))
```

```python
def is_palindrome(s):
   """s нь тэмдэгт мөр
      Хэрэв s палиндром бол True, эсрэг тохиолдолд False буцаана.
       Цэг цэглэл, хоосон зай, том үсгүүдийг тооцохгүй."""
   
   def to_chars(s):
      s = s.lower()
      letters = ''
      for c in s:
        if c in 'абвгдеёжзийклмноөпрстуүфхцчшщъыьэюя':
            letters = letters + c
      return letters

   def is_pal(s):
      print(f'  is_pal функцийг "{s}" утгатай дуудав')
      if len(s) <= 1:
         print('  Суурь нөхцлөөс True буцаав')
         return True
      else:
         answer = s[0] == s[-1] and is_pal(s[1:-1])
         print(' ', s, 'тэмдэгт мөрөнд', answer, 'буцаав')
         return answer
         
   return is_pal(to_chars(s))

print('Шалгах: Харах')
print(is_palindrome('Харах'))
print('Шалгах: Хариах')
print(is_palindrome('Хариах'))

```

```python
import calendar

# Get the calendar for a specific year and month
year = 2025
month = 11  # November

# monthcalendar returns a list of lists, where each inner list represents a week
# Days outside the specified month are represented by 0
cal_matrix = calendar.monthcalendar(year, month)

# Print the calendar matrix
print(f"Calendar for {calendar.month_name[month]} {year}:")
for week in cal_matrix:
    print(week)
```

```python
def find_scienceday(year):
    month = cal.monthcalendar(year, 11)
    if month[0][cal.SUNDAY] != 0:
        scienceday = month[3][cal.SUNDAY]
    else:
        scienceday = month[4][cal.SUNDAY]
    return scienceday
print('2011 онд', 'Шинжлэх ухааны ажилтны өдөр 11-р сарын',
      find_scienceday(2011), '-нд өдөр тохиосон')
```

```python
from datetime import date

def shopping_days(year):
    """year бол тоо >= 1941
       тухайн жилд Наадам болон Шинэ жилийн баярын хооронд хэдэн өдөр 
       байгааг буцаана."""
    if year < 1941:
        raise ValueError("Year must be >= 1941")
    
    # Наадам: 7-р сарын 11
    naadam_date = date(year, 7, 11)
    # Шинэ жил: 1-р сарын 1 (дараагийн жил)
    new_year_date = date(year, 11, 26)
    
    # Тоолох өдөр
    delta = new_year_date - naadam_date
    return delta.days
print('2023 онд Наадам болон Шинэ жилийн баярын хооронд',
      shopping_days(2023), 'өдөр байна.')
```
